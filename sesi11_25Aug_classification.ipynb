## Classification

saat target kita bersifat kategorikal, kita bisa gunakan algoritma classification

- untuk contoh kali ini, kita coba algoritma Logistic Regression
- kita coba dengan data BMW Sales, dari harga mobil, apakah terjual/tidak


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df_car = pd.read_csv("dataset/BMW_Car_Sales_Classification.csv")
```

```python
df_car.head()
```

```python
df_car.info()
```

```python
# kita siapkan data fitur dan target
X = df_car[['Price_USD']]  # fitur

y_text = df_car['Sales_Classification']  # target
```

```python
y_text.value_counts()
```

data target kita berbentuk text

model ML kita dari sklearn tidak bisa handle data text. kita perlu convert data text tersebut jadi numerik. Misalnya Low = 0, High = 1.

kita gunakan teknik `Encoding`


```python
# kita encode data target kita
encoder = LabelEncoder()
y = encoder.fit_transform(y_text)

print(encoder.classes_)
```

0 = High

1 = Low


```python
pd.Series(y).value_counts()
```

data sudah jadi numerik semua, kita bisa lanjut. seharusnya ada step2 seperti:

- data cleaning
- data splitting (train test)
- data scaling
  dan lain lain...


```python
# latih model
model_car_log = LogisticRegression()
model_car_log.fit(X, y)
```

```python
# gunakan model untuk prediksi
# misalnya, apakah mobil 100 000 USD terjual?
model_car_log.predict([[100000]])
```

menurut model kita, mobil di harga USD 100000 punya chance terjual 'low'

tapi model kita akurat/ tidak? di model Classification kita tidak punya RMSE atau angka error, kita punya:

- accuracy score
- confusion matrix
- classification report

kita bandingkan data asli kita "y" dengan hasil prediksi model kita "y_pred"


```python
y_pred = model_car_log.predict(X)
```

```python
print(accuracy_score(y, y_pred), '\n')

print(confusion_matrix(y, y_pred), '\n')

print(classification_report(y, y_pred))
```

accuracy score kita di 0.695 artinya sekitar 70% hasil prediksi model kita akurat, tepat saat dibandingkan data asli kita


confusion matrix menggambarkan untuk setiap kategori, ada berapa hasil prediksi yang tepat, dan ada berapa yang tidak -> masuk kategori lain. di contoh kita, cuma ada 2 kategori jadi matrixnya berukuran 2 x 2

dari confusion matrix kita, model kita 100% prediksi selalu 'low' selalu kategori 1. jadi dari sini sebenarnya kita tau modelnya kurang bagus.


dari classification report kita bisa lihat, per kategori, precision dan recall kategori tersebut

- precision: kalau model bilang kategori ini, dia benar adalah kategori ini (presisi)
- recall: dari data yang kita gunakan, dari semua data point yang masuk kategori ini, berapa % yang berhasil di prediksi adalah kategori ini oleh modelnya
- f1-score: rata2 antara precision dan recall

sementara untuk accuracy berapa hasil prediksi model yang tepat dengan data aslinya
