## Statistics in Pandas

Kita bisa gunakan fungsi bawaan pandas untuk hitung agregasi data

contohnya, rata2 sales

selain itu ada "groupby" untuk mengelompokkan data sebelum di agregasi

contohnya, rata2 sales per segment


```python
# rata2 sales

df_superstore['Sales'].mean()
```

```python
df_superstore.agg({"Sales": "mean"})
```

```python
df_superstore.groupby("Segment").agg({"Sales": "mean"})
```

bahkan untuk beberapa grup dan beberapa agregasi juga bisa

misalnya, per Segment dan per Region, lalu mau lihat "sum" dan "mean"


```python
df_superstore.groupby(["Segment", "Region"]).agg(
    {"Sales": ["mean", "sum"]})
```

```python
# semua kolom, khusus kolom numerik
df_superstore[["Sales", "Quantity", "Discount", "Profit", "Segment", "Region"]
              ].groupby(["Segment", "Region"]).agg(["mean", "sum"])
```

## Query

filter data dengan kondisi

misalnya kita mau ambil data khusus Customer Name "Chad Cunningham"


```python
kondisi1 = df_superstore['Customer Name'] == "Chad Cunningham"

df_chad = df_superstore[kondisi1]
df_chad
```

```python
df_chad.info()
```

```python
# double check data di kolom Customer Name
df_chad['Customer Name'].unique()
```

Untuk cari text yang _wildcard_ tidak exact match, kita bisa pakai ".str.contains()"

- .str.contains() # ada text ini
- .str.startswith() # dimulai dengan
- .str.endswith() # diakhiri dengan


```python
# contoh: semua customer Chad

kondisi2 = df_superstore['Customer Name'].str.contains("Chad")

df_chad2 = df_superstore[kondisi2]

df_chad2['Customer Name'].unique()
```

untuk angka, query nya bisa juga untuk angka tertentu, lebih besar, lebih kecil, atau bukan angka tertentu


Kita juga bisa gunakan lebih dari 1 kondisi untuk filter nya

misalnya, selain Chad kita juga mau ambil semua Aaron


```python
kondisi2 = df_superstore['Customer Name'].str.contains("Chad")
kondisi3 = df_superstore['Customer Name'].str.contains("Aaron")

# | or, artinya salah satu kondisi terpenuhi
df_chadaaron = df_superstore[kondisi2 | kondisi3]

df_chadaaron['Customer Name'].unique()
```

```python
# misalnya selain Chad, dia harus dari Segment "Home Office"
kondisi2 = df_superstore['Customer Name'].str.contains("Chad")
kondisi4 = df_superstore['Segment'] == "Home Office"

# & and, kedua kondisi harus terpenuhi
df_chadhome = df_superstore[kondisi2 & kondisi4]

print(df_chadhome['Customer Name'].unique())
print(df_chadhome['Segment'].unique())
```
