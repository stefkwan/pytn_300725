## Line Chart

Biasanya digunakan untuk:
- melihat tren dari data numerik berdasarkan tanggal atau tahun

## Bar Chart

Pada latihan ini, mari kita buat contoh vertical bar chart jumlah imigran Iceland ke Canada pada tahun 2008-2011.

```python
df_iceland = df_can.loc['Iceland', years]

df_iceland.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')

# arrow
plt.annotate('',
             xy=(32, 70),
             xytext=(28, 20),
             xycoords='data',
             arrowprops=dict(arrowstyle='->',
                             connectionstyle='arc3',
                             color='blue',
                             lw=2)

             )

# text
plt.annotate('2008 - 2011 Financial Crisis',
             xy=(28, 30),
             rotation=72.5,
             va='bottom',
             ha='left')
plt.show()
```

Selanjutnya, mari kita buat horizontal bar plot yang menunjukkan jumlah imigran ke Canada dari top 15 countries, selama periode 1980-2013.

```python
df_can.sort_values(by='Total', ascending=True, inplace=True)
df_top15 = df_can['Total'].tail(15)
df_top15.head()
```

```python
df_top15.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Countries Contributing to the Immigration toCanada between 1980 - 2013')

for index, value in enumerate(df_top15):
  label = format(int(value), ',') # format int with commas
  # posisikan text di akhir bar
  # kurangi 47000 dari x, dan 0.1 dari y
  # untuk memastikan text di dalam bar
  plt.annotate(label,xy=(value - 47000, index - 0.10),color='white')
plt.show()
```

## Pie Chart

Pada latihan ini mari kita gunakan pie chart untuk menjelajahi proporsi (persentase) imigran baru yang dikelompokkan berdasarkan `Continent` untuk seluruh periode waktu dari 1980 hingga 2013

```python
df_continents = df_can.groupby('Continent').sum()
print(type(df_can.groupby('Continent')))
df_continents.drop(['Region','DevName'],axis=1, inplace=True)
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.2f%%',
                            startangle=90,
                            shadow=True,
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.show()
```

Agar visualisasi kita lebih menarik, kita dapat atur explode, color dan legend pada Pie Chart.

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue','lightgreen', 'pink']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]
df_continents['Total'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%',
                            pctdistance=1.12,
                            startangle=120,
                            shadow=True,
                            labels=None,
                            explode=explode_list,
                            colors=colors_list,
                            )

plt.title('Immigration to Canada by Continent [1980 - 2013]')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels=df_continents.index,
           loc='upper left')
plt.show()
```

## bubble chart

Sekarang, mari kita analisis dampak Great Depression di Argentina. Argentina mengalami krisis ekonomi besar dari tahun 1998 hingga 2002, yang menyebabkan pengangguran tinggi, kerusuhan, jatuhnya pemerintahan, dan gagal bayar hutang luar negeri. Lebih dari 50% penduduk hidup dalam kemiskinan, dan pada tahun 2002, tujuh dari sepuluh anak di Argentina hidup dalam kemiskinan.

Untuk memahami dampak krisis ini, mari kita:
- Bandingkan imigrasi Argentina dengan Brasil dari tahun 1980 hingga 2013.
- Gunakan bubble plot dengan bobot gelembung yang dinormalisasi berdasarkan populasi setiap tahun.


```python
df_can_t = df_can[years].transpose()
df_can_t.index.name = 'Year'
df_can_t.reset_index(inplace=True)
df_can_t.head()
```

```python
# normalisasi data Brazil
norm_brazil = (df_can_t['Brazil'] - df_can_t['Brazil'].min()) /(df_can_t['Brazil'].max() - df_can_t['Brazil'].min())
# normalisasi data Argentina
norm_argentina = (df_can_t['Argentina'] -df_can_t['Argentina'].min()) / (df_can_t['Argentina'].max() -df_can_t['Argentina'].min())
```

```python
# Brazil
ax0 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Brazil',
                    figsize=(14, 8),
                    alpha=0.5,
                    color='green',
                    s=norm_brazil * 900 + 10,
                    xlim=(1975, 2015)
                    )
# Argentina
ax1 = df_can_t.plot(kind='scatter',
                    x='Year',
                    y='Argentina',
                    alpha=0.5,
                    color="blue",
                    s=norm_argentina * 900 + 10,
                    ax = ax0
                    )
ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Brazil and Argentina from 1980 - 2013')
ax0.legend(['Brazil', 'Argentina'], loc='upper left', fontsize='x-large')
```
